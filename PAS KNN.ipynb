```python
df=pd.read_csv('/content/drive/MyDrive/XI-PPLG-7/MP-1b/train.csv')
df
```

```python
df = df.dropna()
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for Kelamin, d in df.groupby('Sex'):
    ax.scatter(d['Fare'], d['Age'], label=Kelamin)

plt.legend()
plt.title('Data Kelamin Pengunjung Titanic')
plt.xlabel('Tarif')
plt.ylabel('Umur')
plt.grid(True)
plt.show()
```

```python
import numpy as np

x_train = np.array(df[['Fare', 'Age']])
y_train = np.array(df['Sex'])

print(f'x_train:\n{x_train}\n')
print(f'y_train: {y_train}')
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)

print(f'y_train:\n,{y_train}')
```

```python
y_train = y_train.flatten()
print(f'y_train: {y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(x_train, y_train)
```

```python
tarif = 75.0000
umur = 70

x_new = np.array([[tarif, umur]]).reshape(1, -1)
x_new
```

```python
y_new = model.predict(x_new)
y_new
```

```python
lb.inverse_transform(y_new)
```

```python
# Visualisasi nearest Neighours'
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for Kelamin, d in df.groupby('Sex'):
    ax.scatter(d['Fare'], d['Age'], label=Kelamin)
plt.scatter(
    tarif,
    umur,
    marker='s',
    color='red',
    label='misterius'
)

plt.legend(loc='upper left')
plt.title('Sebaran Data Tarif Usia dan Kelamin')
plt.xlabel('Tarif')
plt.ylabel('Umur')
plt.grid(True)
plt.show()
```

```python
# kalkulasi Distance (Euclidean Distance)
misterius = np.array([tarif, umur])
misterius
```

```python
x_train
```

```python
from scipy.spatial.distance import euclidean

data_jarak = [euclidean(misterius, d) for d in x_train]
data_jarak
```

```python
df.loc[:, 'jarak'] = data_jarak
df.sort_values(['jarak'])
```

```python
# Evaluasi KKN Classification Moel
x_test = np.array([[50.1000, 65], [70.8730, 96], [56, 52], [169, 67]])
y_test = lb.transform(np.array(['1', '1', '0', '1'])).flatten()

print(f'x_test:\n{x_test}\n')
print(f'y_test: {y_test}')
```

```python
# Prediksi terhadap testing set
y_pred = model.predict(x_test)
y_pred
```

```python
# Accurancy
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
print(f'Accurancy: {acc}')
```

```python
from sklearn.metrics import precision_score

prec = precision_score(y_test, y_pred)
print(f'Precision: {prec}')
```

```python
from sklearn.metrics import recall_score

rec = recall_score(y_test, y_pred)
print(f'Recall: {rec}')
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)
print(f'F1-Score: {f1}')
```

```python
from sklearn.metrics import classification_report

cls_report = classification_report(y_test, y_pred)
print(f'Classification Report:\n{cls_report}')
```

```python
from sklearn.metrics import matthews_corrcoef

mcc = matthews_corrcoef(y_test, y_pred)
print(f'MCC: {mcc}')
```
